### Good Loan Metrics

```python
good_loans = df[df['loan_status'].isin(["Fully Paid", "Current"])]
total_loan_applications = df['id'].count()

good_loan_applications = good_loans['id'].count()
good_loan_funded_amount = good_loans['loan_amount'].sum()
good_loan_received_amount = good_loans['total_payment'].sum()

good_loan_funded_amount_millions = good_loan_funded_amount / 1000000
good_loan_received_amount_millions = good_loan_received_amount / 100000

good_loan_percentage = (good_loan_applications / total_loan_applications) * 100

print("Good Loan Applications",good_loan_applications )
print("Good Loan Funded Amount (in million): ${:.2f}M".format(good_loan_funded_amount_millions))
print("Good Loan Received Amount (in million): ${:.2f}M".format(good_loan_received_amount_millions))
print("%age of Good Loan Applications: {:.2f}%".format(good_loan_percentage))
```

### Bad Loan Metrics

```python
bad_loans = df[df['loan_status'].isin(["Charged Off"])]
total_loan_applications = df['id'].count()

bad_loan_applications = bad_loans['id'].count()
bad_loan_funded_amount = bad_loans['loan_amount'].sum()
bad_loan_received_amount = bad_loans['total_payment'].sum()

bad_loan_funded_amount_millions = bad_loan_funded_amount / 1000000
bad_loan_received_amount_millions = bad_loan_received_amount / 100000

bad_loan_percentage = (bad_loan_applications / total_loan_applications) * 100

print("Bad Loan Applications",bad_loan_applications )
print("Bad Loan Funded Amount (in million): ${:.2f}M".format(bad_loan_funded_amount_millions))
print("Bad Loan Received Amount (in million): ${:.2f}M".format(bad_loan_received_amount_millions))
print("%age of Bad Loan Applications: {:.2f}%".format(bad_loan_percentage))
```

### Monthly Trends by Issue date for Total Funded Amount

```python
monthly_funded = (
    df.sort_values('issue_date') #will create cronologocal order for all the months starts from jan-dec
      .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort=False)['loan_amount']
      .sum()
      .div(1_000_000)
      .reset_index(name='loan_amount_millions')
)

plt.figure(figsize=(10, 5)) #plot figure 10 inches wide 5 inches hide

plt.fill_between( #what color we need to fill (Starts an Area Chart)
    monthly_funded['month_name'],   #filling it on month_name and loan_amount _millions
    monthly_funded['loan_amount_millions'],
    color = 'skyblue',
    alpha=0.5 #for color transparency
)

plt.plot( #plotting the line Chart
    monthly_funded['month_name'], #based on the month_name and the total_funded_amount
    monthly_funded['loan_amount_millions'],
    color = 'blue', #color to fill in that line
    linewidth = 2 #width
)

for i, row in monthly_funded.iterrows(): #to add labels in to the particular line chart
    plt.text( #Adds text labels on chart
        i,
        row['loan_amount_millions'] + 0.1,
        f"{row['loan_amount_millions']:.2f}",
        ha='center',
        va='bottom',
        fontsize=9,
        rotation=0,
        color = 'black'
    )

plt.title('Total Funded Amount By Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Funded Amount ($ Million)')
plt.xticks(
    ticks=range(len(monthly_funded)),
    labels=monthly_funded['month_name'],
    rotation=45
)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Monthly Trends By Issue Date For Total Received Amount

```python
monthly_received = (
    df.sort_values('issue_date')
      .assign(month_name = lambda X: X['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort=False)['total_payment']
      .sum()
      .div(1000000)
      .reset_index(name = 'received_amount_millions')
)

plt.figure(figsize = (10, 5))
plt.fill_between(monthly_received['month_name'], 
                 monthly_received['received_amount_millions'],
                 color = 'lightgreen', #color for the area chart
                 alpha = 0.5)
plt.plot(monthly_received['month_name'],
         monthly_received['received_amount_millions'],
         color = 'green', #color for the line chart
         linewidth = 2
        )

for i, row in monthly_received.iterrows():
    plt.text(i, row['received_amount_millions'] + 0.1,
             f"{row['received_amount_millions']:.2f}",
             ha = 'center',
             va = 'bottom',
             fontsize = 9,
             rotation = 0,
             color = 'black',
            )

plt.title('Total Received Amount By Month', fontsize = 14)
plt.xlabel('Month')
plt.ylabel('Received Amount ($ Millions)')
plt.xticks(ticks = range(len(monthly_received)), 
            labels = monthly_received['month_name'], 
            rotation = 45)
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.tight_layout()
plt.show()
```

### Monthly Trends By Total Loan Applications

```python
monthly_applications = (
    df.sort_values('issue_date')
      .assign(month_name = lambda X: X['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort= False,)['id']
      .count()
      .reset_index(name = 'loan_applications_count')
)

plt.figure(figsize = (10, 5))
plt.fill_between(monthly_applications['month_name'],
                 monthly_applications['loan_applications_count'],
                 color = 'yellow',
                 alpha = 0.5
                )
plt.plot(
    monthly_applications['month_name'],
    monthly_applications['loan_applications_count'],
    color = 'brown',
    linewidth = 2
)

for i, row in monthly_applications.iterrows():
    plt.text(
        i, row['loan_applications_count'] + 0.5,
        f"{ row['loan_applications_count']}",
        ha = 'center',
        va = 'bottom',
        fontsize = 9,
        rotation = 0,
        color = 'black'
    )
plt.title('Total Loan Applications By Month', fontsize = 14)
plt.xlabel('Month')
plt.ylabel('Number Of Applications')
plt.xticks(
    ticks= range(len(monthly_applications)),
    labels = monthly_applications['month_name'],
    rotation = 45 
)
plt.grid(True, linestyle='--', alpha = 0.6)
plt.tight_layout()
plt.show()    
    
```

### Regional Analysis By State For Total Funded Amount

```python
state_funding = df.groupby('address_state')['loan_amount'].sum().sort_values(ascending=True)
state_funding_thousands = state_funding / 1000

plt.figure(figsize = (10, 8)) #to plot a bar chart
#barh - horizontal bar chart
bars = plt.barh(state_funding_thousands.index, state_funding_thousands.values, color = 'lightcoral')

for bar in bars:
    width = bar.get_width()
    plt.text(width + 10, bar.get_y() + bar.get_height() / 2,
            f'{width:,.0f}K', va = 'center', fontsize = 9)
plt.title('Total Funded Amount by State {in $ Thousands}')
plt.xlabel('Funded Amount($ \'000)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

### Loan Term Analysis by Total Funded Amount(Donut Chart)

```python
term_funding_millions = df.groupby('term')['loan_amount'].sum()/ 1000000

plt.figure(figsize = (5, 5))
plt.pie(
    term_funding_millions,
    labels = term_funding_millions.index,
    autopct = lambda p: f"{p:.1f}%\n${p*sum(term_funding_millions)/100:.1f}M",
    startangle = 90,
    wedgeprops = {'width': 0.4}
)
plt.gca().add_artist(plt.Circle((0,0),0.70, color = 'white'))
plt.title("Total Funded Amount By Term(in $ Millions)")
plt.show()
```
